```python
df = pd.DataFrame({'col1': list('ABBC'), 'col2': list('ZZXY')})
df["color"] = np.where(df["col2"] == "Z", "green", "red")
conditions = [
    (df['col2'] == 'Z') & (df['col1'] == 'A'),
    (df['col2'] == 'Z') & (df['col1'] == 'B'),
    (df['col1'] == 'B')
]
choices = ["yellow", "blue", "purple"]
df["color"] = np.select(conditions, choices, default="black")
df
```
