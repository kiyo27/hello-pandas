```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
df = pd.DataFrame([[11, 21, 31], [12, 22, 32], [31, 32, 33]],
                  index=['one', 'two', 'three'], columns=['a', 'b', 'c'])
df2 = df[["a", "c"]]

with pd.ExcelWriter("multi.xlsx") as writer:
    df.to_excel(writer, sheet_name="sheet1")
    df2.to_excel(writer, sheet_name="sheet2")
    
```
